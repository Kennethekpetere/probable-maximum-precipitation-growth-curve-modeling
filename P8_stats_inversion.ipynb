```python
import pandas as pd

def compute_inversion_rate(pmp_file: str, inversion_stats_file: str):
    # Step 1: Read the input CSV file
    data = pd.read_csv(pmp_file)

    # Extract the relevant fields
    duration_columns = ['30-min', '1-hour', '2-hour', '3-hour', '6-hour', '12-hour', '24-hour', '48-hour', '72-hour']
    if not all(col in data.columns for col in duration_columns):
        raise ValueError("One or more required columns are missing in the input CSV file.")

    # Define the duration pairs for calculations
    duration_pairs = [
        ('30-min', '1-hour'),
        ('1-hour', '2-hour'),
        ('2-hour', '3-hour'),
        ('3-hour', '6-hour'),
        ('6-hour', '12-hour'),
        ('12-hour', '24-hour'),
        ('24-hour', '48-hour'),
        ('48-hour', '72-hour')
    ]

    # Initialize the output structure
    metrics = {'metric': ['count', 'percentage']}

    for col1, col2 in duration_pairs:
        # Calculate differences between successive durations
        differences = data[col2] - data[col1]

        # Count inversions (negative differences)
        inversion_count = (differences < 0).sum()

        # Calculate the percentage of inversions
        total_count = len(differences)
        inversion_percentage = (inversion_count / total_count) * 100

        # Add results to the metrics dictionary
        column_name = f"{col1} to {col2}"
        metrics[column_name] = [inversion_count, inversion_percentage]

    # Convert metrics dictionary to a DataFrame
    metrics_df = pd.DataFrame(metrics)

    # Save the output to the specified CSV file
    metrics_df.to_csv(inversion_stats_file, index=False)


# # Annual PMP file inputs
# input_file = "annual_max_pmp.csv"
# output_file = "stat_inversion_annual_pmp.csv"

# # Run Function
# compute_inversion_rate(input_file, output_file)
# print("Inversion rates computed and Saved to Ouput file")
```

```python
# Station Annual PMP file inputs
input_file = "annual_max_pmp_stn.csv"
output_file = "stat_inversion_annual_pmp_stn.csv"

# Run Function
compute_inversion_rate(input_file, output_file)
print("Inversion rates computed and Saved to Ouput file")
```
